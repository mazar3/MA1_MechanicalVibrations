### List Basics <a class="anchor" id="list"></a>

Lists are one of the most versatile and commonly used data structures in Python. They allow you to store multiple items in a single variable. Here are some key points about lists:

1. **Creating a List**: Creating a list is as simple as putting different comma-separated values between square brackets.

    ```python
    my_list = [1, 2, 3]
    mixed_list = [1, "Hello", 3.4]
    ```

2. **Accessing Elements**: Accessing elements of a list is done by using the index of the element in the list. The index of the first element is 0, the second element is 1, and so on.

    ```python
    my_list = [1, 'Mechanical', 3]
    print(my_list[0])  # 1
    print(my_list[1])  # Mechanical
    ```

3. **Negative Indexing**: Negative indexing means beginning from the end, -1 refers to the last item, -2 refers to the second last item, and so on.

    ```python
    my_list = [1, 'Mechanical', 3]
    print(my_list[-1])  # 3
    ```

4. **Slicing a List**: Slicing a list is done by using the colon operator.

    ```python
    my_list = [5, 5, 6, 6, 7, 8, 9, 10.5]
    print(my_list[2:5])  # [6, 6, 7]
    print(my_list[:5])   # [5, 5, 6, 6, 7]
    print(my_list[5:])   # [8, 9, 10.5]
    print(my_list[-5:-2])  # [6, 7, 8]
    print(my_list[::2])  # [5, 6, 7, 9]
    print(my_list[::-1])  # [10.5, 9, 8, 7, 6, 6, 5, 5]
    ```

5. **Modifying Elements**: You can modify elements of a list by assigning a new value to a specific index.

    ```python
    my_list = [1, 'Mechanical', 3]
    my_list[1] = 'Vibrations'
    print(my_list)  # [1, 'Vibrations', 3]
    ```

6. **List Methods**: Python lists come with a variety of useful methods.

    ```python
    my_list = [1, 2, 3]
    my_list.append(4)          # [1, 2, 3, 4]
    my_list.insert(1, 5)       # [1, 5, 2, 3, 4]
    my_list.remove(3)          # [1, 5, 2, 4]
    my_list.pop()              # [1, 5, 2]
    print(len([1, 2, 3]))      # 3
    print([1, 2, 3] + [4, 5, 6])  # [1, 2, 3, 4, 5, 6]
    ```

### Generating Arrays with 'numpy as np' using `np.linspace` and `np.arange`
`np.linspace(start, stop, num)` returns `num` evenly spaced samples, calculated over the interval `[start, stop]`. E.g. `np.linspace(0, 10, 5)` returns `array([ 0.,  2.5,  5.,  7.5, 10.])`.

`np.arange(start, stop, step)` returns evenly spaced values within a given interval. E.g. `np.arange(0, 10, 2)` returns `array([0, 2, 4, 6, 8])`.

Note that `np.linspace` includes the stop value, while `np.arange` does not.

**Assignment 1.7** : Complete the function `generate_time_vector` that generates a time vector from `start_time` (included) to `end_time` (not included)  with a sampling frequency of `f`.

````python
import numpy as np
def generate_time_vector(start_time, end_time, f):
    """
    Generates a time vector from start_time to end_time with a step of time_step
    """

    if f == 0 : return np.array([])
    time_step = 1/f
    time_vector = np.arange(start_time, end_time, time_step)
    return time_vector
````

**Assignment 1.9** : Build a function that make a vector rotation of an angle $\theta$ of a vector $\bf{a} \in \mathbb{R}^3$. The rotation is done around the z-axis, so the vector $\bf{a}$ is rotated in the xy-plane. The function should return the rotated vector $\bf{b}$.\
The function should be called `rotate_vector` and get two arguments, the vector $\bf{a}$ and the angle $\theta$ in degrees.\
***TIP*** : Matrix multiplication in numpy is done using `np.dot(A,B)` or `A@B` where A and B are matrices.\
While elementwise multiplication is done using `np.multiply(A,B)` or `A*B` where A and B are arrays or matrices.\
\
The rotation matrix is given by:
$
\begin{bmatrix}
\cos(\theta) & -\sin(\theta) & 0 \\
\sin(\theta) & \cos(\theta) & 0 \\
0 & 0 & 1
\end{bmatrix}
$

````python
def rotate_vector_xy(vector, angle):
    """
    Rotates a 2D vector by angle degrees
    : vector : np.array
    : angle : float in degrees
    """
    angle = np.deg2rad(angle)
    rotation_matrix = np.array([[np.cos(angle), -np.sin(angle), 0],
                                [np.sin(angle), np.cos(angle), 0],
                                [0, 0, 1]])
    rotated_vector = np.dot(rotation_matrix, vector)
    return rotated_vector
````

````python
assert np.isclose(rotate_vector_xy(np.array([1,0,0]), 90), np.array([0,1,0])).all()
assert np.isclose(rotate_vector_xy(np.array([1,0,0]), 180), np.array([-1,0,0])).all()
assert np.isclose(rotate_vector_xy(np.array([1,0,0]), 270), np.array([0,-1,0])).all()
assert np.isclose(rotate_vector_xy(np.array([1,0,0]), 360), np.array([1,0,0])).all()

````

**Assignment 2.1** : Finish the function `find_first_value_greater_than(matrix, a)` which returns the first value (working row by row, [0,0]->[0,1]->...->[0,n]->[1,0]->...->[n-1,n-1] ) in a matrix that is greater than the value `a`.

**BIG Tip** : you can use the `return` command in any point of your function, when `return` is reached the function terminates and returns the corresponding value
